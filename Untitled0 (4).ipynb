3. Navigation Bar with Different Borders

```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Navigation Bar with Different Borders</title>
  <style>
    /* Styling the navigation bar */
    .navbar {
      display: flex;

      background-color: #4a90e2;
      padding: 15px;
    }

    /* Common styling for all links */
    .navbar a {
      color: white;
      text-decoration: none;
      padding: 12px 20px;
      margin: 0 10px;
      font-size: 18px;
    }

    /* Different border styles for each link */
    .home {
      border: 2px solid red;
    }

    .about {
      border: 2px dashed green;
    }

    .services {
      border: 2px dotted blue;
    }

    .blog {
      border: 2px double orange;
    }

    .contact {
      border: 2px groove purple;
    }

    /* Hover effect (optional to look good) */
    .navbar a:hover {
      background-color: #555;
      transition: background-color 0.3s;
    }
  </style>
</head>
<body>

  <!-- Navigation bar -->
  <div class="navbar">
    <a href="#" class="home">Home</a>
    <a href="#" class="about">About Us</a>
    <a href="#" class="services">Services</a>
    <a href="#" class="blog">Blog</a>
    <a href="#" class="contact">Contact</a>
  </div>

</body>
</html>
```
